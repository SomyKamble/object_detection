```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = 'frozen_inference_graph.pb'
```

```python
model = cv2.dnn_DetectionModel(frozen_model,config_file)
model
```

```python
#to get the label names

classlables =[]
file_name = 'coco.names.txt'
with open(file_name,'rt') as fpt:
    classlables = fpt.read().rstrip('\n').split('\n')
    
```

```python
classlables[17]
```

```python
print((classlables))
```

```python
print(len(classlables))
```

```python
img = cv2.imread('cat2.jpg')
```

```python
#model resizinzing 

model.setInputSize(320,320)
model.setInputScale(1.0/127.5)#255/2 =127.5 
model.setInputMean((127.5,127.5,127.5))##mobilenet => [-1,1]
model.setInputSwapRB(True)#autoswaping
```

```python
plt.imshow(img) #bgr
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
ClassIndex , confidence , bbox = model.detect(img,confThreshold=0.5)
```

```python
model.detect(img,confThreshold=0.5)
```

```python
font_scale=3
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd ,conf, boxes in zip(ClassIndex.flatten(),confidence.flatten(),bbox):
    cv2.rectangle(img,boxes,(255,0,1),3)
    cv2.putText(img,classlables[ClassInd-2],(boxes[0]+10,boxes[1]+40),font,fontScale=font_scale,color=(0,255,0),thickness=3)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
#taking the input form the webcam itself


cap = cv2.VideoCapture(0)


if not cap.isOpened():
    cap = cv2.VideoCapture(0)
    

if not cap.isOpened():
    raise IOError("cannot open the video")
    

font_scale = 4

font = cv2.FONT_HERSHEY_PLAIN



while True:
    
    ret,frame = cap.read()
    ClassIndex,confidence,bbox= model.detect(frame,confThreshold=0.5)
    
    print(ClassIndex)
    
    
    if (len(ClassIndex) != 0):
        for ClassInd,conf,boxes in zip(ClassIndex.flatten(),confidence.flatten(),bbox):
            if ClassInd <=80:
                cv2.rectangle(frame,boxes,(255,0,1),3)
                cv2.putText(frame,classlables[ClassInd-1],(boxes[0]+10,boxes[1]+40),font,fontScale=font_scale,color=(0,0,255),thickness=3)
    cv2.imshow("object detection tutorial",frame)
    
    if cv2.waitKey(2) & 0XFF == ord('q'):
        break
    
    
cap.release()
cv2.destroyAllWindows()
            


```
